## By a Recursive  function we can say that a function is going to call itself again and again

```python
## Factorial of a Number
## Factorial of 8 --> 8*7*6*5*4*3*2*1
## Factorial of 98 --> 98*97*96...*3*2*1
## factorial of 0 or factorial of 1 --> 1
```

```python
def Factorial(num):
    if num==0 or num==1:
        print(f'The factorial of {num} is 1')

    else:
        factorial=1
        for i in range(1,num+1):
            factorial=factorial*i
        print(f"The factorial of {num} is {factorial}")
```

```python
from math import factorial


number=int(input("Enter the number: "))

factorial(number)
```

```python
def factorial_recursive(num):
    if num==0 or num==1:
        return 1
    else:
        return num* factorial_recursive(num-1)
```

```python
factorial_recursive(1)
```

```python
factorial_recursive(3)
```

```python
factorial_recursive(5)
```

```python
## return num* factorial_recursive(num-1) 
'''
base condition ---> which will stop the loop or which will let the function to stop calling itself 
in our case , base condition was ---> if i==0 or i==1 --> 1
num* factorial_recursive(num-1) 
num==5
5*factorial_recursive(4) & factorial_recursive(4) ---> 4*factorial_recursive(3)
factorial_recursive(3) --> 3*factorial_recursive(2) 
factorial_recursive(2) --> 2*factorial_recursive(1)
'''
```

```python
input_recursive=int(input("Enter the value: ")) 
print(f"The factorial of the given number {input_recursive} is : ", factorial_recursive(input_recursive))
```

## Funtion with variable length Arguments

## Positional Arguments--> are been used so that we can pass multiple calues inside our function as an argument irrespective of the number odf value we have mentioned as the parameter inside our function

```python
## funtion to add multiple numbers irrespective of how many numbers we are passing

def addition_positional(*num):
    add=0
    for i in num:
        add+=i
    print("Addition of the given numbers are: ",add)
```

```python
addition_positional(1,2)
```

```python
addition_positional(1,2,5,6546,86,513,54,84,65,984,654)
```

## KEyword Arguments --> Take the input as the Keyword values and irrespective of any number of keyword values you are going to pass it wont matter

```python
def data(**user_data):
    for key,value in user_data.items():
        print(f"{key} : {value}")
```

```python
data(name="Pranav",age=25,gender='Male')
```

## You want to pass keyword value pairs as your input along with positional arguments as well

```python
def variable(*positional,**keyword):
    for i in positional:
        print(f'positional value is: {i}')
    for key,value in keyword.items():
        print(f'keyword value is: {key} : {value}')
```

```python
variable(1,2,3,4,5,name="Pranav",age=25,gender='Male')
```
