# Lambda Functions

### Lambda Functions are the One liner Functions defined by lambda keyword

## Syntex for Lambda Functions:

## Lambda arguments: expression

## Lambda--> keyword which simpally let the python interpreter know that the current function is our lambda function

## Argument--> Input values for our function

## Expression ---> Logicc ( Function Body)

### lambda x:x**2

## They can have any number or arguments but only one expression or one logic
## They are used for short Operations or as Arguments to Higher Order Functions

```python
## It can be treated as a function without a Name And they dont even required Return Statement

def square(num):
    return num**2
```

```python
print(square(5))
```

```python
## This similar problem can be solved using LAmbda as well
```

```python
Lambda_function=lambda x:x**2
```

```python
type(Lambda_function)
```

```python
Lambda_function(5)
```

```python
## lambda function to calculate sum of 2 numbers
sum=lambda x,y:x+y
print(sum(5,10))
```

```python
product=lambda a,b,c,d:a*b*c/4
print(product(2,3,4,5))
```
