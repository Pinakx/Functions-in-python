```python
## define a Function that would behave a password strength checker.

'''
Stron password-->
if you password:
    1. length is greater than 8 characters
    2. password contains atleast a lowercase cherecter
    3. password contains atleast an uppercase cherecter
    4. password contains atleast a number
    5. password contains atleast a special character

Weak password-->
If your password donst meet the above criteria, then it is a weak password.
'''

def password_strength_checker(password):
    special_characters = "!@#$%^&*()-+"

    if len(password) < 8:
        return f"Weak password: Too short (minimum 8 characters)"

    if not any(char.islower() for char in password):
        return f"Weak password: Needs a lowercase letter"

    if not any(char.isupper() for char in password):
        return f"Weak password: Needs an uppercase letter"

    if not any(char.isdigit() for char in password):
        return f"Weak password: Needs a number"

    if not any(char in special_characters for char in password):
        return f"Weak password: Needs a special character ({special_characters})"

    return f"Strong password: {password}"
```

```python
password_strength_checker("password123")
```

```python
password_strength_checker("hello")
```

```python
password_strength_checker("HelloAmit@123")
```

# Palindron String
## Palindrom string is going to have characters similar to it from both sides or even if you will reverse the string still you will be getting the given string itself

## palindrome --> aba, aabbaa, aabbcbbaa
## not palindrome --> ab, abA,  acb, 123
##
## reversing a string value--> string[::-1]

```python
## Check that we if the current string is palindrome or not.

def palindrome_checker(string):
    
    string_value=string.lower().replace(" ","")

    if string_value == string_value[::-1]:
        print("The current string is a palindrome")

    else:
        print("The current string is not a palindrome")
```

```python
palindrome_checker('Aba') ##Aba--> aba
```

```python
palindrome_checker("ab") ## ab--> ba not palindrome
```

```python
palindrome_checker('A man a plan a canal Panama')
```
